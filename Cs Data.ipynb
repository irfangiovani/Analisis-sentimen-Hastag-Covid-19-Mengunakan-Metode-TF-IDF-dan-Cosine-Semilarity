```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
import operator
```

```python
dataset = {}
dataset["d1"] = "ppkm sudah tidak diterapkan di indoneisa"
dataset["d2"] = "masyarakat telah dibolehkan tidak pakai masker"
dataset["d3"] = "kota makasar kota"
dataset["d4"] = "ada acara tunjuk ijin satgas covid"
dataset["d5"] = "ada diaspora sebar indonesia nilai menteri bumn dpt salah faktor bantu pulih ekonomi indonesia pasca pandemi covid dengannetmerdekaberdaulat"
dataset["d6"] = "adik sd negeri tanjak iii lawan covid varian harap vaksin bs sekolah"
dataset["d7"] = "ah aja yg lewat mikirnya tuh sdg ngomong ngadepin virus dewa covid"
dataset["d8"] = "airlangga akselerasi vaksinasi disiplin prokes kunci kendali covid sindonews bukanberita"
dataset["d9"] = "ajar guru papar covid diskes pekanbaru gejala ringan"
dataset["d10"] = "aktif covid belu tambah pasien tinggal"
dataset["d11"] = "ampuh tangkal virus covid kuat imun teliti ungkap minum kopi bukti turun risiko tular virus covid"
dataset["d12"] = "anak positif covid mohon isolasi rumah bawa penting pantau anak anak nampak lemah mahu makan nafas laju tunggu lama bawa hospital"
dataset["d13"] = "anak serang flu pilek pandemi covid tangan celebritiesdotid infocelebterkini"
dataset["d14"] = "anda negara yg inisiatif menkreasikan paspor bentuk kartu konten kolaborasi data imigrasi data sehat lawan covid jalan bangsa negara"
dataset["d15"] = "anggap situasi covid persipura jayapura ancam kalah wo"
dataset["d16"] = "anggota babinsa terima harga dedikasi komitmen tangan covid"
dataset["d17"] = "anggota dpr dukung perintah hati ubah status pandemi covid"
dataset["d18"] = "anggota koramil kemusu sana satgas covid kec kemusu melaksanskan awas prokes ptm sd n bawu kec kemusukab boyolali"
dataset["d19"] = "anggota koramil nogosari terima arah bimbing jam komandan koramil kapten inf warisno kait kembang covid wilayah kec nogosari"
dataset["d20"] = "anggota polres situbondo konfirmasi covid varian omicron"
dataset["q"] = "ppkm di makasar gawat covid"
```

```python
#frekuensi kemunculan kata
tf = CountVectorizer()
term_doc_matrix = tf.fit_transform(dataset.values())
pd.DataFrame(term_doc_matrix.toarray(), index=dataset.keys(), columns=tf.get_feature_names())
```

```python
#Pembobotan kata
ed = euclidean_distances(term_doc_matrix, term_doc_matrix)
df_ed =pd.DataFrame(ed, index=dataset.keys(), columns=dataset.keys())
df_ed
```

```python
#Perhitungan Cosin Similarity
rank_ed ={}
for k in dataset.keys():
    if k != "q":
        rank_ed[k] = df_ed.at[k, "q"]
top_rank_ed = dict(sorted(rank_ed.items(), key=operator.itemgetter(1)))
pd.DataFrame(top_rank_ed.values(), index=top_rank_ed.keys(), columns=["Euclidean Distance"])
```

```python
#Penentuan Ringking Cosin Similarity
tfidf = TfidfVectorizer()
inverted_index = tfidf.fit_transform(dataset.values())
pd.DataFrame(inverted_index.toarray(), index=dataset.keys(), columns=tfidf.get_feature_names())
```

```python
#Perhitungan Cosin Similarity 
cs = cosine_similarity(inverted_index, inverted_index)
df_cs = pd.DataFrame(cs, index=dataset.keys(), columns=dataset.keys())
df_cs
```

```python
#pengurutan Rangking Cosin Similarity
rank_cs = {}
for k in dataset.keys():
    if k !="q" :
        rank_cs[k] = df_cs.at[k, "q"]
top_rank_cs =dict(sorted(rank_cs.items(), key=operator.itemgetter(1), reverse=True))
pd.DataFrame(top_rank_cs.values(), index=top_rank_cs.keys(), columns=["cosine Similarity"])
```
